```python
from networkx import Graph, draw_networkx
import networkx as nx
from scipy.io import mmread, mmwrite
import matplotlib.pyplot as plt
import os
```

```python
for size in [100, 500, 1000, 2500, 5000, 10000]:
    # Generate a graph with size nodes, where the first node is connected to all other nodes, and the other nodes have no other outward connections
    G = nx.star_graph(size)
    # Save the graph to a file
    mmwrite(f"graphs/star_{size}.mtx", nx.adjacency_matrix(G))
```
