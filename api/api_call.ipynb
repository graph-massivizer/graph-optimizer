```python
import requests

url = 'http://localhost:5000/predict'
with open('hardware_demo.json', 'r') as hardware, open('input_dag_demo.json', 'r') as bgo_dag, open('test_graph_props.json', 'r') as graph_props:
    form_data = {
        'hardware': hardware.read(),
        'bgo_dag': bgo_dag.read(),
        'graph_props': graph_props.read()
    }

response = requests.post(url, data=form_data)

print(response)

if response.status_code == 200:
    print('POST request successful')
else:
    print('POST request failed')

```
