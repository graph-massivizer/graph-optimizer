```python
import requests
import json

url = 'http://localhost:5000/'
with open('hardware.json', 'r') as hardware, open('input_dag_demo.json', 'r') as bgo_dag, open('test_graph_props.json', 'r') as graph_props:
    form_data = {
        'hardware': hardware.read(),
        'bgo_dag': bgo_dag.read(),
        'graph_props': graph_props.read()
    }
```

```python
models_response = requests.post(url + '/models', data=form_data)

print(models_response.text)
```

```python
evaluate_response = requests.post(url + '/evaluate', data=form_data)

print(evaluate_response.text)
```
